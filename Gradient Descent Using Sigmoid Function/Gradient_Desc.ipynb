```
import numpy as np
import random
import traceback
import math
import matplotlib.pyplot as plt

X = [0.75, 3.24, 5.04, 0.45, 0.48]
Y = [0.61, 0.83, 0.95, 0.15, 0.166]

try:
    def sigmoid_function (w,b,x):
        return 1.0 / (1.0 + np.exp(-(w*x + b)))

    def error (w,b):
        err = 0.0
        for x,y in zip(X,Y):
            fx = sigmoid_function(w,b,x)
            err += 0.5 * (fx - y)** 2
        return err

    def gradient_b(w,b,x,y):
        fx = sigmoid_function(w,b,x)
        return (fx - y) * fx * (1-fx)

    def gradient_w(w,b,x,y):
        fx = sigmoid_function(w,b,x)
        return (fx - y) * fx * (1-fx) * x

    def gradient_descent():
        B,W,E = [],[],[]
        w,b,learning_rate,epochs = 0,0,0.1,1000
        for i in range(epochs):
            dw, db = 0, 0
            for x,y in zip(X,Y):
                dw += gradient_w(w,b,x,y)
                db += gradient_b(w,b,x,y)
            w = w - (learning_rate * dw)
            b = b - (learning_rate * db)
            B.append(b)
            W.append(w)
            E.append(error(w,b))
            print("Weights :",w)
            print("Bias : ",b)
            print("Error",error(w,b))
            print("-------------------")
        plotting(B,W,E)

    def plotting(B,W,E):
        plt.plot(B)
        plt.title("Bias vs Epochs")
        plt.xlabel("Epochs")
        plt.ylabel("Bias")
        plt.show()

        print()
        print()

        plt.plot(W)
        plt.title("Weights vs Epochs")
        plt.xlabel("Epochs")
        plt.ylabel("Weights")
        plt.show()
        
        print()
        print()

        plt.plot(E)
        plt.title("Error vs label")
        plt.xlabel("Epochs")
        plt.ylabel("Error")
        plt.show

    gradient_descent()   
except:
    traceback.print_exc()
```
