```
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
from tensorflow.keras import datasets
from tensorflow.keras.utils import to_categorical
```

```
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()
```

```
print(train_images.shape)
print(test_images.shape)
print(train_labels.shape)
print(test_labels.shape)
```

```
#Reshape

train_images = train_images.reshape((-1,784))
test_images = test_images.reshape((-1,784))

print(train_images.shape)
print(test_images.shape)
```

```
model = Sequential([Dense(400, activation='relu', input_shape=(784,)),Dense(204, activation='relu'),Dense(84, activation='relu'),Dense(40, activation='relu'),Dense(10, activation='softmax')])
model1 = Sequential([Dense(400, activation='relu', input_shape=(784,)),Dense(204, activation='relu'),Dense(84, activation='relu'),Dense(40, activation='relu'),Dense(10, activation='softmax')])
model2 = Sequential([Dense(400, activation='relu', input_shape=(784,)),Dense(204, activation='relu'),Dense(84, activation='relu'),Dense(40, activation='relu'),Dense(10, activation='softmax')])
model3 = Sequential([Dense(400, activation='relu', input_shape=(784,)),Dense(204, activation='relu'),Dense(84, activation='relu'),Dense(40, activation='relu'),Dense(10, activation='softmax')])
model4 = Sequential([Dense(400, activation='relu', input_shape=(784,)),Dense(204, activation='relu'),Dense(84, activation='relu'),Dense(40, activation='relu'),Dense(10, activation='softmax')])
```

```
model.compile(optimizer = 'RMSprop',loss = 'categorical_crossentropy',metrics = ['accuracy'])
model1.compile(optimizer = 'adam',loss = 'categorical_crossentropy',metrics = ['accuracy'])
model2.compile(optimizer = 'adadelta',loss = 'categorical_crossentropy',metrics = ['accuracy'])
model3.compile(optimizer = 'adagrad',loss = 'categorical_crossentropy',metrics = ['accuracy'])
model4.compile(optimizer = 'nadam',loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```
history = model.fit(train_images,to_categorical(train_labels),validation_split = 0.2,epochs = 100,batch_size=32, shuffle=True)
print()
print("**************************************************************************************************************************************")
print()
history1 = model1.fit(train_images,to_categorical(train_labels),validation_split = 0.2,epochs = 100,batch_size=32, shuffle=True)
print()
print("**************************************************************************************************************************************")
print()
history2 = model2.fit(train_images,to_categorical(train_labels),validation_split = 0.2,epochs = 100,batch_size=32, shuffle=True)
print()
print("**************************************************************************************************************************************")
print()
history3 = model3.fit(train_images,to_categorical(train_labels),validation_split = 0.2,epochs = 100,batch_size=32, shuffle=True)
print()
print("**************************************************************************************************************************************")
print()
history4 = model4.fit(train_images,to_categorical(train_labels),validation_split = 0.2,epochs = 100,batch_size=32, shuffle=True)
print()
print("**************************************************************************************************************************************")
print()
```

```
print("RMSprop ------",model.evaluate(test_images,to_categorical(test_labels),verbose = 2))
print("Adam ------",model1.evaluate(test_images,to_categorical(test_labels),verbose = 2))
print("Adadelta ------",model2.evaluate(test_images,to_categorical(test_labels),verbose = 2))
print("Adagrad ------",model3.evaluate(test_images,to_categorical(test_labels),verbose = 2))
print("Nadam ------",model4.evaluate(test_images,to_categorical(test_labels),verbose = 2))
```
